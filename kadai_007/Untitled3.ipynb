```python
# Step1: 必要なライブラリのインポート
import pandas as pd
import matplotlib.pyplot as plt

# Step2: CSVファイルの読み込み
df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv("category.csv")

# Step3: 商品番号でマージしてカテゴリー列を追加
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# Step4: カテゴリーの出現頻度を棒グラフで表示
df['カテゴリー'].value_counts().plot(kind='bar', color='lightblue', edgecolor='gray')
plt.title('カテゴリー別の出現頻度')
plt.xlabel('カテゴリー')
plt.ylabel('件数')
plt.tight_layout()
plt.show()

# Step5: 商品番号ごとの注文数の基本統計量を表示
stats = df.groupby('商品番号')['注文数'].describe()
print(stats)
```
